### Importação das bibliotecas e carregamento da imagem

```python

import matplotlib.pyplot as plt
import rasterio

caminho = '/content/drive/MyDrive/Colab Notebooks/NDVI.tif'
ndvi = rasterio.open(caminho)

```

### Print de metadados para verificar se a imagem foi aberta corretamente

```python
print("Largura:", ndvi.width)
print("Altura:", ndvi.height)
print("Quantidade de bandas:", ndvi.count)
print("Projection:", ndvi.crs)
print("Transform matrix:", ndvi.transform)
```

```python
print(ndvi.read())
```

### Crianção do grafico

```python



#Alocando a banda de cor em uma variavel
banda = ndvi.read(1)

#Criando mapa de para valores positivos e negativos
cmap = plt.cm.RdYlGn

#Configurando a banda com o mapa

fig, ax = plt.subplots(figsize=(10,10))
img = ax.imshow(banda, cmap=cmap, vmin=-1, vmax=1)

#Aparencia do Grafico
cbar = fig.colorbar(img, ax=ax, fraction=0.046, pad=0.04)
cbar.ax.set_ylabel('NDVI', rotation=90)

#Colocando titulo
ax.set_title('NDVI')

plt.show()
```
