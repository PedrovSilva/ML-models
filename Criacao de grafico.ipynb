```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

caminho = '/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/arquivo_final.csv'
sep = ';'
dtype = 'float'
bloco = 19
df = pd.read_csv(caminho, sep=sep, dtype=float)

df.rename(columns={'# BlockId': 'BlockID'}, inplace=True)

df.head()
```

```python
base = df.query('BlockID == @bloco')

media = base.filter(regex='^(Mean|Voo)')
mediana = base.filter(regex='^(Median|Voo)')
```

```python

for coluna in media.columns:
  if coluna != 'Voo':
    plt.plot(media['Voo'], media.loc[:,coluna], label=coluna)

plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.legend()
plt.title(f'Bloco {bloco} Medias')
plt.xlabel('Voo')
plt.ylabel('Valores')

plt.show()
```

```python
for coluna in mediana.columns:
  if coluna != 'Voo':
    plt.plot(mediana['Voo'], mediana.loc[:,coluna], label=coluna)

plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.legend()
plt.title(f'Bloco {bloco} Medianas')
plt.xlabel('Voo')
plt.ylabel('Valores')

plt.show()
```
