Definicao de variaveis e import de bibliotecas

```python
import os
import pandas as pd
import glob


sep=';'
dtype='float'

folders = ["/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/voo1", "/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/voo2", "/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/voo3", '/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/voo4']
folders = ["/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/voo4"]
file_pattern = "*.csv"



```

Pegando block ids

```python
arqbase = pd.read_csv('/content/drive/MyDrive/Faculdade /Projeto CAP/Dados/Indices_voos_blocos_19_e_23_embrapa/Estatisticas/voo1/Voo1_NDVI.csv', usecols=['# BlockId'], sep=sep, dtype=dtype)

arqbase['Voo'] =4

```

Procurando arquivos csv

```python
all_files = []
for folder in folders:
    folder_path = os.path.join(os.getcwd(), folder)
    folder_files = glob.glob(os.path.join(folder_path, file_pattern))
    all_files.extend(folder_files)

```

Percorrendo csvs e selecionando apenas as colunas de Media e mediana

```python
dataframes = []
dataframes.append(arqbase)
for file in all_files:
    df = pd.read_csv(file, usecols=["Mean", "Median"], sep=sep, dtype=dtype)
    filename = os.path.splitext(file)[0]  # remove a extensão do arquivo
    filename_parts = filename.split("_")  # divide o nome do arquivo em partes com base no sublinhado
    name_before_dot = filename_parts[0]  # acessa a primeira parte do nome
    name_after_last_underscore = filename_parts[-1]
    new_colnames = {colname: f"{colname}_{name_after_last_underscore}" for colname in df.columns}
    df = df.rename(columns=new_colnames)
    dataframes.append(df)
```

Concatenando dataframes

```python
merged_df = pd.concat(dataframes, axis=1)
```
